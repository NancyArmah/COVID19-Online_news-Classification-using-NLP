```python
from google.cloud import bigquery
client = bigquery.Client()
dataset_ref = client.dataset("covid19", project="gdelt-bq")
dataset = client.get_dataset(dataset_ref)
tables = list(client.list_tables(dataset))
for table in tables:  
    print(table.table_id)
```

```python
query = """
        SELECT * FROM `gdelt-bq.covid19.onlinenews`
        WHERE DATE(DateTime) BETWEEN "2022-01-01" AND "2022-01-31"
        """
query_job = client.query(query)
events = query_job.to_dataframe()
events
```

```python
events.to_csv("covid_onlinenews2022.csv")
import os
os.chdir(r'../working')
from IPython.display import FileLink
FileLink(r'covid_onlinenews2022.csv')
```
